```python
import pandas as pd
```

```python
gun_violence_pd = pd.read_csv("gun-violence.csv")
gun_violence_pd
```

```python
gun_violence_pd.columns
```

```python
participant_type_pd = gun_violence_pd[["incident_id", "date", "state", "n_killed", "latitude", "longitude", "participant_type"]].copy()
participant_type_pd["participant_type"] = participant_type_pd["participant_type"].str.split(pat="\|\|")
participant_type_pd = participant_type_pd.explode('participant_type').reset_index(drop=True)
participant_type_pd[["participant_id", "participant_type"]] = participant_type_pd.participant_type.str.split("::", expand=True)
participant_type_pd
```

```python
participant_gender_pd = gun_violence_pd[["incident_id", "participant_gender"]].copy()
participant_gender_pd["participant_gender"] = participant_gender_pd["participant_gender"].str.split(pat="\|\|")
participant_gender_pd = participant_gender_pd.explode('participant_gender').reset_index(drop=True)
participant_gender_pd[["participant_id", "participant_gender"]] = participant_gender_pd.participant_gender.str.split("::", expand=True)
participant_gender_pd = participant_gender_pd.drop(participant_gender_pd[participant_gender_pd["participant_gender"] == "Male, female"].index , axis=0)
participant_gender_pd
```

```python
edad_pd = gun_violence_pd[["incident_id", "participant_age"]].copy()
edad_pd["participant_age"] = edad_pd["participant_age"].str.split(pat="\|\|")
edad_pd = edad_pd.explode('participant_age').reset_index(drop=True)
edad_pd[["participant_id", "participant_age"]] = edad_pd.participant_age.str.split("::", expand=True)
edad_pd
```

```python
participant_status_pd = gun_violence_pd[["incident_id", "participant_status"]].copy()
participant_status_pd["participant_status"] = participant_status_pd["participant_status"].str.split(pat="\|\|")
participant_status_pd = participant_status_pd.explode('participant_status').reset_index(drop=True)
participant_status_pd[["participant_id", "participant_status"]] = participant_status_pd.participant_status.str.split("::", expand=True)
participant_status_pd
```

```python
gun_type_pd = gun_violence_pd[["incident_id", "gun_type"]].copy()
gun_type_pd["gun_type"] = gun_type_pd["gun_type"].str.split(pat="\|\|")
gun_type_pd = gun_type_pd.explode('gun_type').reset_index(drop=True)
gun_type_pd[["participant_id", "gun_type"]] = gun_type_pd.gun_type.str.split("::", expand=True)
gun_type_pd
```

```python
data_pd = participant_type_pd.merge(participant_gender_pd, on=['incident_id', 'participant_id'])
data_pd = data_pd.merge(edad_pd, on=['incident_id', 'participant_id'])
data_pd = data_pd.merge(participant_status_pd, on=['incident_id', 'participant_id'])
data_pd = data_pd.merge(gun_type_pd, on=['incident_id', 'participant_id'])
data_pd
```

```python
data_pd.loc[data_pd['participant_status'].str.contains('Injured', na=False), 'participant_status'] = 'Injured'
data_pd.loc[data_pd['participant_status'].str.contains('Killed', na=False), 'participant_status'] = 'Killed'
data_pd.loc[data_pd['participant_status'].str.contains('Unharmed', na=False), 'participant_status'] = 'Unharmed'
data_pd.loc[data_pd['participant_status'].str.contains('Arrested', na=False), 'participant_status'] = 'Unharmed'
data_pd[["incident_id", "participant_status"]].groupby("participant_status").count()
```

```python
data_pd["participant_age"] = data_pd["participant_age"].fillna(0)
data_pd["participant_age"] = data_pd.participant_age.astype('int')
data_pd = data_pd[data_pd.participant_age < 116]
data_pd
```

```python
west  = ('Washington', "Oregon", "California",  "Arizona", "Nevada", "Utah","Idaho", "Montana", "Wyoming", "Colorado", "Alaska","Hawaii" )
north  = ("North Dakota", "Illinois", "Minnesota", "South Dakota", "Nebraska", "Kansas", "Iowa","Wisconsin", "Michigan","Missouri", "Indiana", "Kentucky", "Tennessee")
east  = ("Rhode Island", "Ohio", "Pennsylvania", "New York","Vermont", "Maine", "New Hampshire", "Massachusetts", "Connecticut", "New Jersey", "Delaware", "Maryland", "South Carolina", "North Carolina", "Virginia", "West Virginia")
south  = ("Florida", "Texas","Georgia", "Alabama", "Mississippi","Arkansas", "Louisiana", "Oklahoma", "New Mexico")

def f(row):
    if row['state'] in west:
        return "west"
    if row['state'] in north:
        return "north"
    if row['state'] in east:
        return "east"
    if row['state'] in south:
        return "south"
    return None

data_pd['region'] = data_pd.apply(f, axis=1)
data_pd
```

```python
data_pd.to_csv("data.csv")
```
