```python
import pandas as pd
```

# pop_states_million

```python
pop_states_pd = pd.read_csv("pop_states.csv")
pop_states_pd["n_killed_millon"] = pop_states_pd["n_killed"]*1000000/pop_states_pd["POPESTIMATE2019"]
pop_states_pd
```

```python
pop_states_pd.corr()
```

```python
print("media:", pop_states_pd["n_killed_millon"].mean())
print("median:", pop_states_pd["n_killed_millon"].median())
```

```python
pop_states_pd.to_csv("pop_states_millon.csv")
```

# ownership_data

```python
ownership_HFR_pd = pd.read_csv("ownership.csv", sep=";")
ownership_HFR_pd["HFR"] = ownership_HFR_pd["HFR"].str.replace(",", ".").astype("float")
ownership_HFR_pd = ownership_HFR_pd[ownership_HFR_pd["Year"]>=2013][["STATE", "HFR"]].groupby("STATE").mean().reset_index()
ownership_HFR_pd
```

```python
ownership_GFI_pd = pd.read_csv("ownership.csv", sep=";")
ownership_GFI_pd["GFI"] = ownership_GFI_pd["GFI"].str.replace(",", ".").astype("float")
ownership_GFI_pd = ownership_GFI_pd[ownership_GFI_pd["Year"]>=2013][["STATE", "GFI"]].groupby("STATE").mean().reset_index()
ownership_GFI_pd
```

```python
data_pd = pd.read_csv("data.csv")
data_pd = data_pd[["state", "incident_id"]].groupby("state").nunique()
data_pd = data_pd.reset_index().rename(columns={"state":"STATE"})
data_pd
```

```python
data_pd = data_pd.merge(ownership_HFR_pd, on=["STATE"])
data_pd = data_pd.merge(ownership_GFI_pd, on=["STATE"])
data_pd
```

```python
data_pd.corr()
```

```python
data_pd.to_csv("ownership_data.csv")
```
