```python
import pandas as pd
```

```python
data_pd = pd.read_csv("data.csv")
data_pd
```

```python
data_pd = data_pd[data_pd.participant_age > 0]
data_pd
```

```python
data_pd = data_pd[["incident_id", "participant_age", "participant_type"]]
data_pd
```

```python
hist = data_pd[data_pd["participant_type"] == "Subject-Suspect"].groupby("participant_age").count().reset_index()
hist = hist.rename({"incident_id":"count"}, axis=1).drop(columns="participant_type")
hist
```

```python
hist.std()
```

```python
desviacion_tipica = 1705.57976
for n in range(1, len(hist)):
    if abs(hist["count"][n-1] - hist["count"][n]) > desviacion_tipica:
        print(hist["participant_age"][n])
```
