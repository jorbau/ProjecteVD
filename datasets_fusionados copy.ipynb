```python
import pandas as pd

elecciones = pd.read_csv('elecciones.csv')
armas = pd.read_csv('gun.csv')
```

```python
elecciones
```

```python
armas = armas.sort_values("date")
armas
```

```python
armas["electoral_year"] = ""
armas.loc[armas["date"] <= "2016-09-08", "electoral_year"] = 2012
armas.loc[armas["date"] > "2016-09-08", "electoral_year"] = 2016
armas
```

```python
def year(x):
    return x[:4]
armas = armas[["date", "state","n_killed", "n_injured", "electoral_year"]]
print(armas.isna().sum())
armas["date"] = armas["date"].apply(year)
armas_year = armas.rename(columns={"date":"year"})
elecciones = elecciones.rename(columns={"year":"electoral_year"})
armas_year
```

```python
armas_year.columns
armas_year
```

```python
armas_year_grouped = armas_year.groupby(["state", "electoral_year"]).n_injured.agg("count").reset_index()
armas_year_grouped
```

```python
def low(x):
    return  x.lower()

armas_year_grouped["state"] = armas_year_grouped["state"].apply(low)
elecciones["state"] = elecciones["state"].apply(low)
#armas_year_grouped = armas_year_grouped.astype({"year": int}, errors='raise')
armas_year_grouped
```

```python
print(armas_year_grouped.electoral_year.unique())
print(elecciones.electoral_year.unique())
```

```python
elecciones
```

```python
combined = pd.merge(armas_year_grouped, elecciones)
```

```python
combined
```

```python
combined = combined.rename(columns={"n_injured":"incidents"})
combined
```

```python
combined.loc[combined["electoral_year"] == 2012, "electoral_year"] = "2012-2016"
combined.loc[combined["electoral_year"] == 2016, "electoral_year"] = "2016-2020"
combined
```

```python
combined.loc[combined["electoral_year"] == "2012-2016", "incidents"] /=3 
combined.loc[combined["electoral_year"] == "2016-2020", "incidents"] /=2 
```

```python
combined.to_csv("partidos_violencia_alternativo.csv", index=False)
```
