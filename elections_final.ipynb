```python
import pandas as pd
import matplotlib.pyplot as plot
```

```python
dataset = pd.read_csv("1976-2020-president.csv")
dataset
```

```python
cleandf = dataset.groupby(["state",  "year", "party_simplified"]).agg({'candidatevotes':'max'})
cleandf
```

```python
cleandf = cleandf.reset_index().groupby(["state", "year", "party_simplified"]).candidatevotes.agg("max").reset_index()
cleandf.party_simplified.agg("max")
```

```python
cleandf.to_csv('clean_elections.csv', index=True)
```

```python
data = pd.read_csv('clean_elections.csv')
data
# data.groupby(["state", "year"])['candidatevotes'].max()
```

```python
votes_maxes = data.groupby(["state", "year"]).candidatevotes.transform(max)
cleandf =  data[data.candidatevotes == votes_maxes]
cleandf.rename(columns={"party_simplified":"party", "candidatevotes":"votes"}, inplace=True)
cleandf
```

```python
cleandf.to_csv("elecciones.csv", index=False)
```
